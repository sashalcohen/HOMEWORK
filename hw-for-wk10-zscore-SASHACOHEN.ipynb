Which albums stood out (AS IN THE BIGGEST)in terms of sales?

```python
## import libraries and packages
import pandas as pd

## mad math functions package
from scipy.stats import zscore 


## run the display code here
pd.options.display.float_format = '{:,.2f}'.format

lps = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/1967_hits.csv")
streams = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/main/most-streamed-2023.csv")
```

```python
lps
```

```python
streams
```

```python
## they are switched
## the only way we can compare is by getting standard deviation in relation to the mean 

lps["sales_zscore"] = zscore(lps["sales"])
lps.sort_values(by="sales_zscore", ascending = False).head(10)
```

```python
streams["streams_2023"] = zscore(streams["streams_2023"])
streams
```
